# now divide the dataset into target variable and input variable then preprocessing to labelEncoding as follow.

```python
inputs=df.drop("salary_more_than_1000k",axis='columns')# assigning new dataframe called inputs for this line and droping
                                                    .# salary_more_than_1000k.
```

```python
target=df['salary_more_than_1000k']
inputs  # our input variables
```

```python
target # our target what we will going to predict by taking input value.
```

# now we are going to import labelEncoder from sklearn inorder to change string value into numbre(for the column names under company,job and Degree)

```python
!pip install preprocessing
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_Degree=LabelEncoder()
```

# creating extra column in dataFrame by transforming as follow in order to preprocessing data  columns into LabelEncoder as follow.

```python
inputs['company_n']=le_company.fit_transform(inputs['company'])
inputs['job_n']=le_job.fit_transform(inputs['job'])
inputs['Dgree_n']=le_Degree.fit_transform(inputs['Degree'])
inputs
```

# In above case company is ecoded by number '2',ABC parma by '0' and '1' for facebook.
# bachers by encoded by 0 and masters by 1.
# sales excutive encoded by 2,business manager encoded by 0 and computer programmer by 1.
# now we drop columns company,job and Degree as follow inorder bring into our model.

```python
inputs_n=inputs.drop(inputs[['company','job','Degree']], axis='columns')# creating the new dataFrame which we called inputs_n.
inputs_n
```

```python
result = pd.concat([inputs_n, target], axis=1) # mixing new dataFrame which contain inputs_n and target variable and assigning 
                                               # result for it.
result
```

# now everything is ok and we can proceed to our model prediction and calling the type of model which we are going to model our data with for determining the performance of our ML model.


```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier() # since our model is DecisionTreeClassifier we put the varible name model.
```

# now we are going to train our model(which is DecisionTreeClassifier in this case) 

```python
model.fit(inputs_n,target)
```

# now our model is already trained. and check accuracy of our model as follow.

```python
model.score(inputs_n,target)
```

# 100/100 our model is will predict for the inputs as follow since it's accuracy is 1.0.

# check how it will predict for given inputs as follow(by giving inputs to it) how   it will predict the target value).


```python
model.predict([[2,0,1]])
```
